```python
import numpy as np
import scipy.stats as stats
import pandas as pd
import matplotlib.pyplot as plt

# Setting a random seed for reproducibility
np.random.seed(42)

# Creating two sample datasets
data = np.random.normal(loc=50, scale=10, size=100)
data2 = np.random.normal(loc=55, scale=10, size=100)
```

```python
# Display the first few rows of the data
df = pd.DataFrame({'Sample 1': data, 'Sample 2': data2})
print("First few rows of the dataset:\n", df.head())
```

```python
# One-sample t-test
# Hypothesis: The mean of the sample is equal to 50
t_statistic, p_value = stats.ttest_1samp(data, 50)
print(f"\nOne-sample t-test\nT-statistic: {t_statistic:.4f}, P-value: {p_value:.4f}")
```

```python
# Visualizing the data distribution for one-sample t-test
plt.figure(figsize=(10, 6))
plt.hist(data, bins=20, alpha=0.5, label='Sample 1')
plt.axvline(np.mean(data), color='blue', linestyle='dashed', linewidth=1)
plt.axvline(50, color='red', linestyle='dashed', linewidth=1)
plt.title('Histogram of Sample 1 Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Two-sample t-test
# Hypothesis : The means of two samples are equal
t_statistic, p_value = stats.ttest_ind(data, data2)
print(f"\nTwo-sample t-test\nT-statistic: {t_statistic:.4f}, P-value: {p_value:.4f}")
```

```python
# Visualizing the data distribution for two-sample t-test
plt.figure(figsize=(10, 6))
plt.hist(data, bins=20, alpha=0.5, label='Sample 1')
plt.hist(data2, bins=20, alpha=0.5, label='Sample 2')
plt.axvline(np.mean(data), color='blue', linestyle='dashed', linewidth=1)
plt.axvline(np.mean(data2), color='orange', linestyle='dashed', linewidth=1)
plt.title('Histogram of Sample Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Paired t-test
# Hypothesis: The mean difference between paired observations is zero
# For paired t-test, we need paired observations. Here, we'll use the first 50 elements of each sample.
t_statistic, p_value = stats.ttest_rel(data[:50], data2[:50])
print(f"\nPaired t-test\nT-statistic: {t_statistic:.4f}, P-value: {p_value:.4f}")
```

```python
# Visualizing the differences for paired t-test
differences = data[:50] - data2[:50]
plt.figure(figsize=(10, 6))
plt.hist(differences, bins=20, alpha=0.5, label='Differences (Sample 1 - Sample 2)')
plt.axvline(np.mean(differences), color='green', linestyle='dashed', linewidth=1)
plt.title('Histogram of Differences for Paired t-test')
plt.xlabel('Difference Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Chi-square test
# Creating a contingency table
observed = np.array([[10, 20, 30], [6, 9, 17]])
chi2, p_value, dof, expected = stats.chi2_contingency(observed)
print(f"\nChi-square test\nChi2: {chi2:.4f}, P-value: {p_value:.4f}, Degrees of Freedom: {dof}")
print("Expected Frequencies:\n", expected)
```

```python
# Visualizing the contingency table
fig, ax = plt.subplots(figsize=(8, 6))
cax = ax.matshow(observed, cmap='coolwarm')
plt.title('Contingency Table')
plt.xlabel('Category')
plt.ylabel('Group')
fig.colorbar(cax)
plt.show()
```
