```python
#Data Cleaning
duplicate_row = df[df.duplicated()]
print(f"Number of Duplicate Rows{len(duplicate_row)}")#number of duplicate
print("*"*100+"\n")

print(duplicate_row)#print row
```

```python
# Remove duplicate value
df.drop_duplicates(inplace = True)
df.shape
```

```python
df.info()
```

```python
# Conver int data into float
df["Type"] = df["Type"].astype(float)
df.head()
```

```python
# Normalizing Data (Example: Scaling scores between 0 and 1)
df["Type"] = df["Type"]/10
df
```

```python
# Handling missing value
# Check for missing values in each column
Column_missing_value = df.isnull().sum()
print("*"*100)
print(Column_missing_value)

# Check for missing values in each rows
row_missing_value = df[df.isnull().any(axis=1)]
print("*"*100)
print(row_missing_value)

# Check Missing value in a specific column
missing_values_in_column = df['Compactness'].isnull().sum()
print("*"*100)
print('Missing values in Compactness are :',missing_values_in_column)
```

```python
# Drop all missing values
df.dropna(inplace = True)
df.isnull().sum()# 9 rows deleted because missing value 
df.shape
```
