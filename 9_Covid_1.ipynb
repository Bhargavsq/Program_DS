# Create a table with data on missing values


```python
# Calculate the total count of missing values for each column
missing_values = df.isnull().sum()

# Calculate the percentage of missing values for each column relative to the total number of rows
percent_missing = df.isnull().sum() / df.shape[0] * 100

# Create a dictionary 'value' containing two key-value pairs:
# - 'missing_values': This key stores the count of missing values for each column.
# - 'percent_missing': This key stores the percentage of missing values for each column.
value = {
    'missing_values ': missing_values,
    'percent_missing %': percent_missing
}

# Create a DataFrame 'frame' using the 'value' dictionary, which combines the count and percentage of missing values.
frame = pd.DataFrame(value)
frame
```

# Create a countplot to visualize the distribution of the 'COVID-19' variable


```python
# Use Seaborn to create a countplot
ax = sns.countplot(x='COVID-19', data=df, legend=False)

# Loop through each bar in the countplot
for p in ax.patches:
    
    # Annotate each bar with the count of occurrences
    ax.annotate(f'\n{p.get_height()}', (p.get_x() + 0.4, p.get_height() + 100), ha='center', va='top', color='white', size=10)

# Show the countplot
plt.show()
```

# Create a pie chart to visualize the distribution of covid-19 cases

```python
# Use the 'value_counts()' method to count the occurrences of each unique value in the 'COVID-19' column
df["COVID-19"].value_counts().plot.pie(explode=[0.1, 0.1], autopct='%1.1f%%', shadow=True)

# Set the title of the pie chart
plt.title('Covid Positive')

# Show the pie chart
plt.show()
```
