```python
import pandas as pd
df = pd.read_csv('E:/MY_WORK/WEEK_2/seedsMMM.csv')
df.head()
```

```python
# Convert to number
df['Area'] = pd.to_numeric(df['Area'], errors='coerce')

# df = df.fillna(df.mean())
# df = df.fillna(df.median())
# df = df.fillna(df.mode().iloc[0])
# df

print("*"*100+"\n")

print('Mean of Area is :',df['Area'].mean())
print('Median of Perimeter is :',df['Perimeter'].median())
print('Median of Compactness is :',df['Compactness'].mode(),'\n')

print("*"*100+"\n")

print('Mode of Asymmetry.Coeff is :',df['Asymmetry.Coeff'].mode().iloc[0])
print('Mode of Kernel.Groove is :',df['Kernel.Groove'].mode().iloc[0],'\n')

print("*"*100+"\n")

fill_values = {
                'Area': df['Area'].mean(), 
                'Perimeter': df['Perimeter'].median(),
                'Compactness': df['Compactness'].mode().iloc[0]
              }
df = df.fillna(fill_values)
```

```python
# Replacing with previous value – Forward fill

df = pd.read_csv('E:/MY_WORK/WEEK_2/seedsMMM.csv')
df = df.ffill()
df.head()
```

```python
# Replacing with next value – Backward fill

df = pd.read_csv('E:/MY_WORK/WEEK_2/seedsMMM.csv')
df = df.bfill()
df.head()
```

```python
# Interpolation ??

df = pd.read_csv('E:/MY_WORK/WEEK_2/seedsMMM.csv')
df = df.interpolate(method='linear')
df.head()
```
